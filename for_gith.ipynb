```python
# Морской бой
#import snoop 

import random
field = [ ["О"] * 6  for _ in range(6) ]
ships = ["■■■", "■■","■■", "■", "■", "■", "■"]

def board():
    #игровой стол
    print( "  | 1 | 2 | 3 | 4 | 5 | 6 |")
    for i in range(6):
        print(f"{i+1}" + "".join([' | ' + str(i) for i in field[i]]) +' |' )

class Ships:
    count = 0
    time_back = "stop"
    
    def __init__(self):
        #распечатка всего флота игрока
        pass
        #self.show = print(" ".join(map(lambda x: f"{x[0]}-> {x[1]}", list(zip(range(1,1+len(ships)),ships)))))
    
    def all_ships(self):
        print( " ".join(map(lambda x: f"{x[0]}-> {x[1]}", list(zip(range(1,1+len(ships)),ships)))))
    #@snoop
    def fight(self,field_1,field_2):
        #Начало сражения по готовым полям
        count = 0
        while True:
            
            if True:
                gun = input(f"выстрел игрока {count%2+1}.Ввести 2 координаты:").split()
                print(gun)
                if len(gun) == 2:
                    if gun[0].isdigit() and gun[1].isdigit() : #len(gun) == 2
                        gun = list(map(int, gun))
                        print( "пред конец")
                        if 1 <= gun[0] <= 6 and 1 <= gun[1] <= 6 and field_1[gun[0]-1][gun[1]-1] != "*":
                            if count%2: #Пк
                                print( "Пальба!")
                                field_1[gun[0]-1] [gun[1]-1] = "*"
                                count += 1
                                board()
                                #бесконечно.Заменить после тестов
                            else: #Чел
                                
                                print( "Пальба!")
                                field_2[gun[0]-1] [gun[1]-1] = "*"
                                count += 1
                                board()
                                #бесконечно.Заменить после тестов
                        else:
                            continue
                else:
                    continue
    
    def ship(self, player):
        #выставление корабля. player: "hum" или "pc"
        def round_ship(y1,x1):
            #Пустая зона вокруг корабля
            if (y1-1) >= 0:
                field[y1-1] [x1] = ' '
            if (y1+1) <= 5:
                field[y1+1] [x1] = ' '
            if (x1-1) >= 0:
                field[y1] [x1-1] = ' '
            if (x1+1) <= 5:
                field[y1] [x1+1] = ' '
            if (y1-1) >= 0 and (x1-1) >= 0:
                field[y1-1] [x1-1] = ' '
            if (y1-1) >= 0 and (x1+1) <= 5:
                field[y1-1] [x1+1] = ' '
            if (y1+1) <= 5 and (x1+1) <= 5:
                field[y1+1] [x1+1] = ' '
            if (y1+1) <= 5 and (x1-1) >= 0:
                field[y1+1] [x1-1] = ' '

        while len(ships)!=0:
            #цикл выставления кораблей
            if player == "hum":
                board()
            if player == "pc":
                boart = self.count
                #board() #Выключить после тестов
            if player == "hum":
                self.all_ships()
            while True:
                try:
                    #print("таймбек до цикла в 58", self.time_back)
                    if self.time_back == "go":
                        #print("timeback")
                        self.time_back = "stop"
                    elif player == "hum":
                        boart = int(input("Выбор корабля. Цифрой: "))
                        boart += -1
                        if ships[boart] not in ships or boart < 0:
                            raise ValueError
                    elif player == "pc":
                        boart = self.count #?????
                    break
                except ValueError as e:
                    print(" ValueError Неверное значение. Выбор корабля. Цифрой:")
                    continue
                except IndexError as e:
                    print("IndexError Неверное значение. Выбор корабля. Цифрой:")
                    continue
                    
            while True:
                try:
                    if player == "hum":
                        #Человек-игрок
                        place = input("Выбор координат корабля гор-но/верт-но.4 Цифры: 2 цифры-начало(раздельно)+"\
                              "2 цифры-конец (раздельно): ").split()
                        y1, x1, y2, x2 = map(lambda x: int(x)-1 , place)
                        #long = len(list(filter(lambda x: 1 <= x <= 5, map(int ,place))))
                    elif player == "pc":
                        #Компьютер игрок
                        #boart = self.count #???
                        y1,x1 = [ random.randint(0,5) for _ in range(2)]
                        rand = random.choice(["гор","вер"])
                        modul = len(ships[self.count]) - 1
                        if rand == "гор":
                            y2 = y1
                            x2 = x1
                            x2 += modul
                            #print("гор", y1+1, x1+1, y2+1, x2+1)
                        if rand == "вер":
                            x2 = x1
                            y2 = y1
                            y2 += modul
                            #print("верт", y1+1, x1+1, y2+1, x2+1)
                                            #5  1  8  1 ????
                        if y1>5 or x1>5 or y2>5 or x2>5:
                            #print("пересчет")
                            continue
                    if y1>y2:
                        y1,y2 = y2,y1
                    if x1>x2:
                        x1,x2 = x2,x1
                        print(y1, x1, y2, x2)
                    if y1 == x1 == y2 == x2:
                        break
                    #print("Сортировка по нарастанию", y1, x1, y2, x2 )
                        #if y1 == x1:
                            #raise ValueError
                    break
                except ValueError:
                    print("Ошибка занчения!")
                except IndexError:
                    continue
                
            boart_long = abs(x2-x1)+1 if abs(x2-x1) > 0 else abs(y2-y1)+1
            #print("1 тест boart_long= ",boart_long)
            #print(ships[boart],boart_long) #Вкл. потом!
            #if player == "pc":
                #boart_long -= 1
            if len(ships[boart]) == boart_long:
                #print("1.1 тест")
                for n in range(boart_long):
                    #print(1,field[y1] [x1+n])
                    #print(2,field[y1+n] [x1])
                    if x1 == x2:
                        if (y1+n) > 5: # or (y2+n) > 5 Если 
                            print("Выход за игровое поле по Y")
                            break
                        if field[y1+n] [x1] != "О":
                            if player == "hum":
                                print("Выберите свободное место, оно занято")
                             #//print(field[y1] [x1+n])
                            self.time_back = "go"
                            break
                    if y1 == y2:
                        if (x1+n) > 5: #or (x2+n) >  5
                            print("Выход за игровое поле по X")
                            break
                        if field[y1] [x1+n] != "О":
                            if player == "hum":
                                print("Выберите свободное место, оно занято")
                             #//print(field[y1] [x1+n])
                            self.time_back = "go"
                            break
                    if x1 == x2:
                        if (y1+n) > 5: # or (y2+n) > 5 Если 
                            print("Выход за игровое поле по Y")
                            break
                    if y1 == y2:
                        if (x1+n) > 5: #or (x2+n) >  5
                            print("Выход за игровое поле по X")
                            break
                    #print("2 тест. С каунт/Борт/Число кораблей", self.count,boart, len(ships))
                    if (abs(y2-y1) == 0) and (abs(x2-x1) > 0) or (abs(x2-x1) == 0 and 
                        abs(y2-y1) == 0) :
                        #Горизонтально
                        for i in range(boart_long):
                            round_ship(y1,x1+i)
                        for i in range(boart_long):
                            field[y1] [x1+i] = '■'
                        #board()
                    elif (abs(y2-y1) > 0) and (abs(x2-x1) == 0):
                        #Вертикально
                        for i in range(boart_long): #Пустые клетки вокруг
                            round_ship(y1+i,x1)
                        for i in range(boart_long): #Расположение кораблей
                            field[y1+i] [x1] = '■'
                        #board()
                    #self.time_back = "stop"
                    #print("Конец кода таймбек", self.time_back, "ships=", ships)
                    #print("ships[boart] = ",ships[boart])
                    del ships[boart]
                    boart += 1
                    
                    if len(ships) == 0:
                        
                        print("успешно")
                        board()
                        return field
                    full = 0
                    for i in field:
                        if "О" not in i:
                            full += 1
                    if full == 6:
                        boart = 7
                        print("нет мест")
                        field = [ ["О"] * 6  for _ in range(6) ]
                        ships = ["■■■", "■■","■■", "■", "■", "■", "■"]
                        board()
                        break
                            
t = Ships()
a = t.ship("pc")
#b = t.ship("pc")
#b = t.ship("hum")
#print(a,b)
t.fight(a,a)
#print(a)
```
